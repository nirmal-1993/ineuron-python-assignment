#### 8. Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam.

```python
def check(spam):
    if spam == 1:
        return 'Hello'
    elif spam == 2:
        return 'Howdy'
    else:
        return 'Greetings!'
    
print(f'if spam=1 then {check(1)}'); print(f'if spam=2 then {check(2)}'); print(f'if spam is Anything else then {check(3)}')
```
